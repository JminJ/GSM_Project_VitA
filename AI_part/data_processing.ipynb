```
def mfcc(fir_file):
  y, sr = librosa.load(fir_file)

  D = np.abs(librosa.stft(y, n_fft = 4096, win_length = 4096, hop_length = 2028))
  mfcc = librosa.feature.mfcc(S = librosa.power_to_db(D), sr = sr, n_mfcc = 20)

  delta2_mfcc = librosa.feature.delta(mfcc, order=2)

  return delta2_mfcc
```

```
def make_show_save(delta2_mfcc, fir_file):
  # download_path = '/content/drive/MyDrive/GSM_Festival_2020_datas'
  download_path = '/content/datas'
  fig = plt.figure(figsize=(12, 4))
  plt.axis('off')

  x = 320 / fig.dpi
  y = 240 / fig.dpi

  fig.set_figwidth(x)
  fig.set_figheight(y)

  fig1 = plt.gcf()
  librosa.display.specshow(delta2_mfcc)

  for i in range(10):
    file_list_D = os.listdir(download_path)

    fir_class = fir_file[28:].split('-')
    class_id = int(fir_class[1])
    print(fir_class)
    if class_id == i:
      this_list_length = len(os.listdir(download_path+'/'+file_list_D[i])) - 1
      fig = plt.figure(figsize=(12, 4))
      plt.axis('off')

      x = 320 / fig.dpi
      y = 240 / fig.dpi

      fig.set_figwidth(x)
      fig.set_figheight(y)

      fig1 = plt.gcf()
      librosa.display.specshow(delta2_mfcc)

      fig1.savefig(download_path+'/'+file_list_D[i]+'/'+file_list_D[i]+'_'+str(this_list_length), bbox_inches='tight', transparent=True, pad_inches = 0)
      plt.close('all')
      print('Complted : '+fir_file[28:])
    else:
      pass
```

```
for_nums = [3, 5, 7, 9]
path = '/content/drive/MyDrive/audio'
a = 0
for i in for_nums:
  file_list = os.listdir(path)

  path_now = path+'/'+file_list[i]
  print(file_list[i])
  file_list = os.listdir(path_now)

  print('--------------------'+path_now+'----------------------')

  for j in file_list:
    a += 1
    print(a)
    fir_file = path_now+'/'+j
    print(fir_file)
    # print(len(fir_file))
    delta2_mfcc = mfcc(fir_file)
    make_show_save(delta2_mfcc, fir_file)
```
